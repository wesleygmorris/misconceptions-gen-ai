```python
import pandas as pd
DATA = './data/'
```

```python
df = pd.read_csv(DATA+'AllItems_AllFinalCodes_split-adjusted_forKaggle_withQuestText.csv')
# Replace NaN values in 'Misconception_Code' with the values from 'Category'
df['Misconception_Code'] = df['Misconception_Code'].fillna(df['Category'])
print(df.columns)
df.head()
```

```python
df['Category'].value_counts()
```

```python
train_ids = [33472, 32833, 31774, 31777, 33471, 31772, 31778, 91695]
validate_ids = [33474, 32829, 104665, 76870]
test_ids = [32835, 109465, 89443]

columns_to_keep = ['QuestionId', 'Question.Text', 'MC_Answer', 'StudentExplanation', 'Misconception_Code']
```

```python
train_df = df[df['QuestionId'].isin(train_ids)].reset_index(drop=True)[columns_to_keep]
validate_df = df[df['QuestionId'].isin(validate_ids)].reset_index(drop=True)[columns_to_keep]
test_df  = df[df['QuestionId'].isin(test_ids)].reset_index(drop=True)[columns_to_keep]
```

```python
print(f"Train:    {len(train_df):,} rows | {train_df['QuestionId'].nunique()} items")
print(f"Validate: {len(validate_df):,} rows | {validate_df['QuestionId'].nunique()} items")
print(f"Test:     {len(test_df):,} rows | {test_df['QuestionId'].nunique()} items")
```

```python
# Check for any rows that didn't land in any split
assigned = set(train_ids + validate_ids + test_ids)
unassigned = df[~df['QuestionId'].isin(assigned)]
if len(unassigned) > 0:
    print(f"\n⚠ {len(unassigned):,} rows with unrecognized QuestionIds:")
    print(unassigned['QuestionId'].value_counts())
else:
    print("\n✓ All rows assigned")
```

```python
print('#####################################')
print('# Questions in the training dataset #')
print('#####################################')

for i in train_df['Question.Text'].unique():
    print(i,'\n')

print('\n\n')

print('#######################################')
print('# Questions in the validation dataset #')
print('#######################################')

for i in validate_df['Question.Text'].unique():
    print(i,'\n')

print('\n\n')

print('#################################')
print('# Questions in the test dataset #')
print('#################################')

for i in test_df['Question.Text'].unique():
    print(i,'\n')
```

```python
print('\n##########################################')
print('# Misconceptions in the training dataset #')
print('##########################################')
print(train_df['Misconception_Code'].value_counts())

print('\n############################################')
print('# Misconceptions in the validation dataset #')
print('############################################')
print(validate_df['Misconception_Code'].value_counts())

print('\n######################################')
print('# Misconceptions in the test dataset #')
print('######################################')
print(test_df['Misconception_Code'].value_counts())
```

```python
train_df['split']='train'
validate_df['split']='valid'
test_df['split']='test'

final_df = pd.concat([train_df, validate_df, test_df])
final_df
```

```python
final_df.to_csv(DATA+'full_df_with_partitions.csv')
```
